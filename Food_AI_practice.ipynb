```python
# Wine 데이터셋 가져오기
wine_quality = fetch_ucirepo(id=186)
X_wine = wine_quality.data.features
y_wine = wine_quality.data.targets
df_wine = pd.concat([X_wine, y_wine], axis=1)
# Rice 데이터셋 가져오기
rice = fetch_ucirepo(id=545)
X_rice = rice.data.features
y_rice = rice.data.targets
df_rice = pd.concat([X_rice, y_rice], axis=1)
print("Wine shape:", df_wine.shape)
print("Rice shape:", df_rice.shape)
```

```python
# Wine
wine_label = "quality"
wine_features = [c for c in df_wine.columns if c != wine_label]
# Rice
rice_label = "Class"
rice_features = [c for c in df_rice.columns if c != rice_label]
print("[Wine]")
print("label:", wine_label)
print("feature 개수:", len(wine_features))
print("feature 목록:", wine_features)
print("\n[Rice]")
print("label:", rice_label)
print("feature 개수:", len(rice_features))
print("feature 목록:", rice_features)
```

```python
def explain_tabular_dataset(df, label_col, dataset_name):
    feature_cols = [c for c in df.columns if c != label_col]
    print(f"\n### {dataset_name} ###")
    print("- sample: 데이터프레임의 한 행(row)")
    print("- label:", label_col)
    print("- feature 개수:", len(feature_cols))
    print("- feature 목록:", feature_cols)
```

```python
# 함수 호출하여 데이터셋 설명 출력
explain_tabular_dataset(df_wine, wine_label, "Wine Quality")
explain_tabular_dataset(df_rice, rice_label, "Rice Cammeo and Osmancik")
```

```python
explain_tabular_dataset(df_wine, "quality", "Wine Quality")
explain_tabular_dataset(df_rice, "Class", "Rice (Cammeo and Osmancik)")
```
